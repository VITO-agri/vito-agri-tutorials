```python
import requests

def list_gcs_files_rest(bucket_name, prefix=""):
    """Lists files in a Google Cloud Storage bucket using REST API"""
    url = f"https://www.googleapis.com/storage/v1/b/{bucket_name}/o"
    params = {"prefix": prefix}
    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        for item in data.get("items", []):
            print(item["name"])
    else:
        print("Error:", response.text)

list_gcs_files_rest("fao-gismgr-asis-data", "DATA/ASIS/MAPSET/DI-D")
```
